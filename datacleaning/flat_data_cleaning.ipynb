# price

```python
train_df['price'].isnull().sum()
```

```python
train_df['price'].value_counts()
```

the price contains 3 type of data lac ,crore and ask for price 

```python
train_df[train_df['price']=='Price on Request']
```

```python
train_df=train_df[train_df['price']!='Price on Request']
```

```python
train_df.shape
```

```python
train_df.sample(4)
```

```python
def convert(a):
    if type(a)==float:
        return a
    else:
        if a[1]=='Lac':
            return round(float(a[0])/100,2)
        else:
            return round(float(a[0]),2)
train_df['price']=train_df['price'].str.split(" ").apply(convert)
```
